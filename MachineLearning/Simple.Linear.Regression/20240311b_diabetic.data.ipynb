```python
# Importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
# Cargar el conjunto de datos de Diabetes
diabetes = datasets.load_diabetes()

# Usar solo una característica
X = diabetes.data[:, np.newaxis, 2]

# Dividir los datos en conjuntos de entrenamiento/prueba 
X_train, X_test, y_train, y_test = train_test_split(
  X, 
  diabetes.target, 
  test_size=0.2, 
  random_state=0
)

# Crear el objeto de regresión lineal
regr = LinearRegression()

# Entrenar el modelo usando los conjuntos de entrenamiento
regr.fit(X_train, y_train)

# Realizar predicciones usando el conjunto de prueba 
y_pred = regr.predict(X_test)
```

```python
# Imprimimos los Coeficientes
print( 'Pendiente: \n', regr.coef_[0] )
print( 'Intercepto: \n', regr.intercept_ )

# Imprimimos las métricas
print( 'Error cuadrático medio: %.2f' % mean_squared_error(y_test, y_pred) )
print( 'Coeficiente de determinación: %.2f' % r2_score(y_test, y_pred) )
```

```python
# Graficamos las salidas 
plt.scatter( X_test, y_test, color = 'black', label = 'Datos Reales' )
plt.plot( X_test, y_pred, color = 'blue', linewidth = 3, label = 'Regresión Lineal' )
plt.xlabel( 'Característica' )
plt.ylabel( 'Progresión de la enfermedad' )
plt.title( ' Regresión Lineal Simple con el Conjunto de Datos de Diabetes' )
plt.legend()
plt.show()
```
