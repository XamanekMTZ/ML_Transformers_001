```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generamos datos sintéticos
np.random.seed(0) # Fijamos la semilla para reproducibilidad
x = np.random.rand(100, 1) * 10 # Generamos 100 valores aleatorios entre 0 y 10
y = 2 * x + 1 + np.random.randn(100, 1 ) * 2 # Generamos y a partir de x con un poco de ruido gaussiano (media 0, desviación 2)
```

```python
# Crear el modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo
model.fit(x, y)

# Hacer predicciones
y_pred = model.predict(x)
```

```python
# Coeficientes
print( 'Pendiente: ', model.coef_[0] )
print( 'Intercepto: ', model.intercept_ )
```

```python
# Métricas
print( 'Error cuadrático medio: %.2f' % mean_squared_error(y, y_pred) )
print( 'Coeficiente de determinación: %.2f' % r2_score(y, y_pred) )
```

```python
# Graficamos
plt.scatter(x, y, color='black', label = 'Datos')
plt.plot(x, y_pred, color='blue', linewidth=3, label = 'Regresión lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión Lineal Simple con Datos Sintéticos')
plt.legend()
plt.show()
```
